# CÓDIGO 1 - USADO EN LA APP DE LOVABLE

```python
extras = ["Bacon", "Queso", "Huevo", "Pepinillos", "N/A"]

compra_hamburguesa = input("¿Quieres comprar una hamburguesa? (S/N)")

if compra_hamburguesa.lower() == "s":
  compra_hamburguesa = True
  print("¿Qué extras quieres con tu hamburguesa?")

  print(" -> " + ", ".join(extras))

  eleccion = input("Extra elegido: ")

  if eleccion in extras:
    print(f"\nPerfecto tu hamburguesa con {eleccion} estará lista en breve.")
  else:
    print(f"\nLo siento, '{eleccion}' no está en nuestra lista de extras.")

else:
  print(f"No hay problema, ¡vuelve cuando tengas hambre!")

```

# APLICANDO APUNTES

# TUPLA

```python
producto_fijo = ("Hamburguesa Especial", 8.50)
stock = 100

print(f"Tenemos nuestra {producto_fijo[0]} a un precio de {producto_fijo[1]} €")

# Un cliente compra 1

pedido = int(input("¿Cuántas hamburguesas quieres comprar? "))

if pedido <= stock :
  stock = stock - pedido # Calcula cuantas nos van a quedar despues de comprar
  print(f"Perfecto ¡PEDIDO CONFIRMADO!, nos quedan {stock} hamburguesas")
else:
  print("No hay suficientes hamburguesas")
```

```python
def calcular_precio(cantidad, precio_unidad):
  total = cantidad * precio_unidad
  return total

precio_final = calcular_precio(5, 8.50)
print(f"El total de su compra es: {precio_final} €")
```
