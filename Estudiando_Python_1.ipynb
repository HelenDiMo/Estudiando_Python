1. El símbolo + (Concatenación) - Se usa para sumar cosas.

    Si son números, los suma matemáticamente ($5 + 5 = 10$).
    
    Si son textos (llamados strings), los "pega" uno al lado del otro.

    ¿Por qué lo usamos? Porque es muy intuitivo.

    Es como decir: "A este trozo de texto añádele este otro".
    
    El problema: Solo puedes usar + con textos.
    
    Si intentas hacer "Edad: " + 25, Python te dará un error porque no sabe cómo "sumar" una palabra y un número.

    

2. El uso de {} (f-strings) - La forma moderna
  
    En realidad, en Python moderno, los programadores preferimos usar llaves {}.
    
    Se llaman f-strings.
    
    Mira la diferencia:
    
      Con + :
        
        print("Hola " + nombre + ", bienvenida")

      Con {} :
      
        print(f"Hola {nombre}, bienvenida")
   
    Nota: Fíjate en esa f que va antes de las comillas. Esa f le dice a Python: "Atención, lo que haya dentro de las llaves es una variable, no un texto".
    
    Es mucho más limpio y evita errores si tienes que mezclar números y palabras.
    
  3. El uso de () (Paréntesis) - Los paréntesis en Python no sirven para "pegar" texto, sino para "llamar" a la acción.
  
      Se usan con las funciones.
    
      Por ejemplo, print() es una función (una acción: imprimir en pantalla).
    
      Lo que va dentro de los paréntesis es el "paquete" que le envías a esa acción para que trabaje con él.

```python
print("Hola " + nombre + ", bienvenida")
print(f"Hola {nombre}, bienvenida")
```
