1. El símbolo + (Concatenación) - Se usa para sumar cosas.

    Si son números, los suma matemáticamente ($5 + 5 = 10$).
    
    Si son textos (llamados strings), los "pega" uno al lado del otro.

    ¿Por qué lo usamos? Porque es muy intuitivo.

    Es como decir: "A este trozo de texto añádele este otro".
    
    El problema: Solo puedes usar + con textos.
    
    Si intentas hacer "Edad: " + 25, Python te dará un error porque no sabe cómo "sumar" una palabra y un número.

    

2. El uso de {} (f-strings) - La forma moderna
  
    En realidad, en Python moderno, los programadores preferimos usar llaves {}.
    
    Se llaman f-strings.
    
    Mira la diferencia:
    
      Con + :
        
        print("Hola " + nombre + ", bienvenida")

      Con {} :
      
        print(f"Hola {nombre}, bienvenida")
   
    Nota: Fíjate en esa f que va antes de las comillas. Esa f le dice a Python: "Atención, lo que haya dentro de las llaves es una variable, no un texto".
    
    Es mucho más limpio y evita errores si tienes que mezclar números y palabras.
    
  3. El uso de () (Paréntesis) - Los paréntesis en Python no sirven para "pegar" texto, sino para "llamar" a la acción.
  
      Se usan con las funciones.
    
      Por ejemplo, print() es una función (una acción: imprimir en pantalla).
    
      Lo que va dentro de los paréntesis es el "paquete" que le envías a esa acción para que trabaje con él.

```python
print("Hola " + nombre + ", bienvenida")
print(f"Hola {nombre}, bienvenida")
```

# Listas



```python
#Lista - Colecciones de objetos que representan secuencias ordenadas de objetos de distintos tipos .Son mutables(pueden alterarse durante la ejecución de un programa)
lista_de_frutas = ["manzana", "banana", "cereza"]
print(lista_de_frutas) # ['manzana', 'banana', 'cereza']
print()
# Para imprimir esta lista de manera simple y bonita:
print(*lista_de_frutas) # manzana banana cereza
```

# Tuplas

```python
#Tupla .- Tienen orden. Pueden contener elementos de distintos tipos. Son inmutables, es decir, no pueden alterarse durante la ejecución de un programa. Se usan habitualmente para representar colecciones de datos una determinada estructura semántica, como por ejemplo un vector o una matriz.
tupla = (1, "dos", 3)
print(tupla) # (1, 'dos', 3)
print()

# Desempaquetado
mi_tupla = (1, "dos", 3)
id_pedido, cantidad_texto, prioridad = mi_tupla
print(f"El pedido {id_pedido}, tiene prioridad {prioridad} y el estado es: {cantidad_texto}") # El pedido 1, tiene prioridad 3 y el estado es: dos
print()
print(*mi_tupla) # 1 dos 3

# Si la tupla es muy larga
print("\n" + "-"*30 + "\n") # Truco para crear una línea divisoria

print("--- LISTA DE DATOS PROCESADOS ---")
for elemento in mi_tupla:
  print(f"Dato procesado: {elemento}")

print("\n" + "-"*30 + "\n") # Truco para crear una línea divisoria

print("--- LISTA DE DATOS SECOS ---")
for elemento in mi_tupla:
  print(elemento)
```

# Diccionarios

```python
DICCIONARIO = {'manzana': 'roja', 'plátano': 'amarillo', 'cereza': 'roja'}
print(f"El plátano es de color {DICCIONARIO['plátano']}")
print()

print("FRUTAS DISPONIBLES:")
for fruta in DICCIONARIO:
  print(f"- {fruta}")
print()

for fruta, color in DICCIONARIO.items():
  print(f"La fruta {fruta} es de color {color}")
```
