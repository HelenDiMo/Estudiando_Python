# TUPLA

```python
producto_fijo = ("Hamburguesa Especial", 8.50)
stock = 100

print(f"Tenemos nuestra {producto_fijo[0]} a un precio de {producto_fijo[1]} €")

# Un cliente compra 1

pedido = int(input("¿Cuántas hamburguesas quieres comprar? "))

if pedido <= stock :
  stock = stock - pedido # Calcula cuantas nos van a quedar despues de comprar
  print(f"Perfecto ¡PEDIDO CONFIRMADO!, nos quedan {stock} hamburguesas")
else:
  print("No hay suficientes hamburguesas")
```
