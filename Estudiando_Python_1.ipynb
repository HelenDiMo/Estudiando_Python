1. El símbolo + (Concatenación) - Se usa para sumar cosas.

    Si son números, los suma matemáticamente ($5 + 5 = 10$).
    
    Si son textos (llamados strings), los "pega" uno al lado del otro.

    ¿Por qué lo usamos? Porque es muy intuitivo.

    Es como decir: "A este trozo de texto añádele este otro".
    
    El problema: Solo puedes usar + con textos.
    
    Si intentas hacer "Edad: " + 25, Python te dará un error porque no sabe cómo "sumar" una palabra y un número.

    

2. El uso de {} (f-strings) - La forma moderna
  
    En realidad, en Python moderno, los programadores preferimos usar llaves {}.
    
    Se llaman f-strings.
    
    Mira la diferencia:
    
      Con + :
        
        print("Hola " + nombre + ", bienvenida")

      Con {} :
      
        print(f"Hola {nombre}, bienvenida")
   
    Nota: Fíjate en esa f que va antes de las comillas. Esa f le dice a Python: "Atención, lo que haya dentro de las llaves es una variable, no un texto".
    
    Es mucho más limpio y evita errores si tienes que mezclar números y palabras.
    
  3. El uso de () (Paréntesis) - Los paréntesis en Python no sirven para "pegar" texto, sino para "llamar" a la acción.
  
      Se usan con las funciones.
    
      Por ejemplo, print() es una función (una acción: imprimir en pantalla).
    
      Lo que va dentro de los paréntesis es el "paquete" que le envías a esa acción para que trabaje con él.

```python
# Diferencia entre usar "print" y "return"

def evaluacion(nota):
    if nota > 5:
        return "Has aprobado" # Devuelve el texto, no lo imprime todavía
    elif nota == 5:
        return "Has aprobado por los pelos"
    else:
        return "Has suspendido"

x = int(input("Nota: "))
print(evaluacion(x)) # El print recibe el texto y lo muestra. ¡Sin None!

# Esta es mejor opción que usar "print" porque así puedes usar ese resultado para otras cosas (como guardarlo en una base de datos o enviarlo por email), no solo para verlo en la pantalla.
```

# BUCLE FOR Y WHILE

```python
# 1. El Bucle for (El Recolector)
# El for se usa cuando sabes exactamente cuántas veces vas a repetir algo o tienes una lista de cosas que recorrer. Es como un recolector que va árbol por árbol en una fila.

# Ejemplo: Tienes una lista de invitados y quieres saludar a cada uno.

invitados = ["Helena", "León", "María Luisa", "Valentín"] # Lista

for persona in invitados:
  print(f"Hola {persona}, Bienvenid@!")
```

```python
# Si queremos que algo se repita X veces usamos range() - empieza a contar desde 0

range(5)

for numero in range(5):
  print(numero)

print("-----------------------------------------------")
print("-----------------------------------------------")

# Si no queeremos empezar de 0 y queremos hacerlo del 1 al 5:
range(1, 5)

for numero in range(1, 6): # El número final tiene que ser donde se detiene, pero sin incluirlo
  print(numero)
```

```python
# 2. El Bucle while (El Vigilante)
# El while se usa cuando no sabes cuántas veces se va a repetir algo. Solo sabes que debe seguir funcionando mientras una condición sea verdad. Es como un vigilante que no se va de la puerta hasta que se haga de día.

i = 0
while i < 5:
  print(i)
  i += 1 # Esto es importante!!! es una forma abreviada (un atajo) de escribir esto: i = i + 1 - sin estyo nuestro bucle estaría dando vueltas todo el rato
```

# RETO: CREAR UN INVENTARIO

Eres el gerente de una tienda de zapatos y necesitas controlar tu inventario.

Cada vez que vendes un par de zapatos, disminuyes tu
inventario.

Quieres tener un programa que te permita ingresar la cantidad inicial de pares de zapatos en tu tienda y luego, sucesivamente, te pregunte cuántos pares de zapatos se vendieron.

El programa debe continuar preguntando cuántos pares de zapatos se vendieron hasta que el inventario llegue a cero.

Finalmente, el programa debe decirte cuántas transacciones (ventas individuales) tuvieron lugar antes de que se acabara el inventario.

Instrucciones:

● Inicia con un inventario de X pares de zapatos (donde X es un número que ingresas).

● Pregunta al usuario cuántos pares de zapatos se vendieron.

● Deduce esa cantidad del inventario.

● Repite el proceso hasta que el inventario sea 0.

● Imprime el número total de transacciones realizadas.


```python
inventario = 100
transacciones = 0

print(f"Nuestro inventario tiene {inventario} zapatos")
while inventario > 0:
  if inventario <= 10:
    print(f"INVENTARIO PRÓXIMO A AGOTARSE, NOS QUEDAN {inventario} PARES DE ZAPATOS")

  num_zapatos = int(input(f"¿Cuántos zapatos se han vendido hoy?"))
  transacciones += 1

  if num_zapatos > inventario:
    print(f"No nos quedan tantos pares, sólo nos quedan {inventario} pares.")
  else:
    inventario -= num_zapatos

    print(f"Nos quedan {inventario} zapatos")

  if inventario == 0:
    print("Se nos han acabado los zapatos")


print(f"Hoy se han realizado {transacciones} ventas")


```

# PROYECTO

El programa generará un número aleatorio entre 1 y 100, y el usuario tendrá que adivinarlo.

El programa dará pistas si el número es
mayor o menor hasta que el usuario acierte.


```python
import random

def juego_adivinanza ():
  numero_secreto = random.randint(1, 100)
  intentos = 9
  adivinado = False

  print("¡Bienvenido al juego de adivinanza de números!")
  print("Estoy pensando un número entre 1 y 100")

  while not adivinado:
    experimento = int(input("¿Cuál crees que es el número? "))
    intentos += 1

    if experimento < numero_secreto:
      print("El número es mayor")
    elif experimento > numero_secreto:
      print("El número es menor")
    else:
      adivinado = True
      print(f"¡Felicidades! Adivinaste el número en {intentos} intentos")

juego_adivinanza()
```

# RETO 3

Dependiendo de la hora, tu app te dirá si es
la hora de:
- Desayuno
- Comida
- Merienda
- Cena
Por cada tiempo de comida te ofrece 2 o 3
platos a elegir



```python
from datetime import datetime
import pytz

hora = datetime.now().hour

menu_saludable = {
    "Desayuno": ["Avena con frutas", "Tostada con aguacate", "Yogur natural"],
    "Comida": ["Pollo a la plancha con arroz", "Ensalada de garbanzos", "Crema de calabacín"],
    "Merienda": ["Frutos secos", "Una manzana", "Batido de proteínas"],
    "Cena": ["Pescado al horno", "Tortilla de espinacas", "Sopa de verduras"]
}

if 6 <= hora < 11:
  momento = "Desayuno"
elif 12 <= hora < 16:
  momento = "Comida"
elif 17 <= hora < 19:
  momento = "Merienda"
elif 20 <= hora < 24:
  momento = "Cena"
else:
  momento = "No es momento de comer"

print(f"Son las {hora}:00. Te toca: {momento}")

if momento != "No es momento de comer":
  opciones = menu_saludable[momento]
  print(f"Opciones disponibles")
  for i, plato in enumerate(opciones, 1):
    print(f"{i}. {plato}")

  eleccion = int(input("\n¿Qué plato prefieres hoy? (1/2/3): "))
  print(f"¡Genial! Has elegido {opciones[eleccion-1]}. ¡Que aproveche!")

else:
    print("Ahora mismo lo mejor es descansar o beber un poco de agua.")
```
