Añadimos:
* while
* capitalize - para que no importe como escribe el cliente el nombre de las burgers o los extras
* quitamos la variable [N/A] de los extras porque ya tenemos la opcion de decir que NO los queremos

```python
inventario = 5
Burgers = ["Simple", "Doble", "Chicken"]
precios = {"Simple": 5, "Doble": 7, "Chicken": 7.5}
extras = ["Bacon", "Queso", "Huevo", "Pepinillos"]

def aplicar_descuento(precio):
  return precio - 1

while inventario > 0:
  print(f"\n--- Nueva Venta (Stock actual: {inventario}) ---")
  print(f"\nBienvenid@ a Helena's Burguer!")

  compra_hamburguesa = input("¿Quieres comprar una hamburguesa? (S/N)")

  if compra_hamburguesa.lower() == "s":
    cantidad = int(input(f"¿Cuántas quieres? (Máximo {inventario}): "))

    if cantidad <= inventario:
      print("¿Qué hamburguesa quieres?")
      for burger in Burgers:
        print(burger)

      eleccion_burger = input("Burger elegida: ").capitalize() # Para que no importe si el cliente lo escribe con mayuscula o minúscula

      add_extras = input("¿Quieres añadir extras? (S/N)")

      if add_extras.lower() == "s":
        print(" -> " + ", ".join(extras))
        eleccion_extra = input("Extra elegido: ").capitalize()
      else:
        print(f"Perfecto, no añadiremos extras a tu hamburguesa.")
      precio_final = aplicar_descuento(precios[eleccion_burger]*cantidad)
      inventario = inventario - cantidad # Para que se "cierre el bucle" una vez hayamos llegado a 0

      print(f"\nEl precio total es {precio_final} €, con 1€ de decuento por Hamburguesa")
      print(f"\nQuedan {inventario} hamburguesas en el inventario.")
    else:
      print(f"Lo siento, no tengo tantas. Solo me quedan {inventario}.")



  else:
    print(f"No hay problema, ¡vuelve cuando tengas hambre!")
    break

if inventario == 0:
  print("\n--- ¡AGOTADO! Hemos cerrado por falta de stock ---")
```

AÑADIMOS:

¿Te gustaría que viéramos cómo hacer que el precio cambie si el cliente elige un extra con coste? (Por ejemplo, que el bacon sume 0.50€).


```python
inventario = 5
Burgers = ["Simple", "Doble", "Chicken"]
precios = {"Simple": 5, "Doble": 7, "Chicken": 7.5}
extras = ["Bacon", "Queso", "Huevo", "Pepinillos", "N/A"]
precios_extras = {"Bacon": 0.5, "Queso": 0.5, "Huevo": 1.0, "Pepinillos": 0.2, "N/A": 0}

def aplicar_descuento(precio):
  return precio - 1

while inventario > 0:
  print(f"\n--- Nueva Venta (Stock actual: {inventario}) ---")
  print(f"\nBienvenid@ a Helena's Burguer!")

  compra_hamburguesa = input("¿Quieres comprar una hamburguesa? (S/N)")

  if compra_hamburguesa.lower() == "s":
    cantidad = int(input(f"¿Cuántas quieres? (Máximo {inventario}): "))

    if cantidad <= inventario:
      print("¿Qué hamburguesa quieres?")
      for burger in Burgers:
        print(burger)

      eleccion_burger = input("Burger elegida: ").capitalize() # Para que no importe si el cliente lo escribe con mayuscula o minúscula

      add_extras = input("¿Quieres añadir extras? (S/N)")

      if add_extras.lower() == "s":
        print(" -> " + ", ".join(extras))
        eleccion_extra = input("Extra elegido: ").capitalize()

        if eleccion_extra in precios_extras:
          coste_extra = precios_extras[eleccion_extra]
        else:
          print("Ese extra no existe, no se aplicará coste adicional")
      else:
        print(f"Perfecto, no añadiremos extras a tu hamburguesa.")

      precio_unidad = aplicar_descuento(precios[eleccion_burger] + coste_extra)
      precio_final = precio_unidad * cantidad

      inventario = inventario - cantidad # Para que se "cierre el bucle" una vez hayamos llegado a 0

      print(f"\nEl precio total es {precio_final} €, con 1€ de decuento por Hamburguesa")
      print(f"\nQuedan {inventario} hamburguesas en el inventario.")
    else:
      print(f"Lo siento, no tengo tantas. Solo me quedan {inventario}.")



  else:
    print(f"No hay problema, ¡vuelve cuando tengas hambre!")
    break

if inventario == 0:
  print("\n--- ¡AGOTADO! Hemos cerrado por falta de stock ---")
```
